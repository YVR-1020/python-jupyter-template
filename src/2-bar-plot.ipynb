### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 12, 39] # Example values, you can adjust these

# Define a list of colors for each bar
# You can use named colors, hex codes, or RGB tuples
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum']

# Create the bar plot
plt.bar(categories, values, color=colors)

# Add a title to the plot
plt.title("Bar Plot with Custom Colors")

# Add labels for the axes (optional but good practice)
plt.xlabel("Categories")
plt.ylabel("Values")

# Show the plot
plt.show()
```
