### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the scatter plot as provided in the image
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10] # Completed based on the pattern in the image

# Create the scatter plot
plt.scatter(x, y)

# Add annotations to each point
for i, (xi, yi) in enumerate(zip(x, y)):
    # Annotate with (x, y) value slightly offset from the point
    plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(0,10), ha='center')

# Set the title of the plot
plt.title("Scatter Plot with Annotations") # Completed title based on the image

# Add labels for the axes (optional but good practice)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add a grid for better readability (optional)
plt.grid(True)

# Show the plot
plt.show()
```
