### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Data for the plots ---

# Line plot data
x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)

# Bar chart data
categories_bar = ['A', 'B', 'C', 'D']
values_bar = [15, 30, 20, 40]
colors_bar = ['skyblue', 'lightcoral', 'lightgreen', 'gold']

# Scatter plot data
x_scatter = np.random.rand(50) * 10
y_scatter = np.random.rand(50) * 10
sizes_scatter = np.random.rand(50) * 200 + 50 # Random sizes for points
colors_scatter = np.random.rand(50) # Random colors for points

# Pie chart data
labels_pie = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes_pie = [25, 30, 20, 25] # Must sum to 100 or represent proportions
colors_pie = ['lightgrey', 'lightsalmon', 'lightsteelblue', 'darkseagreen']


# --- Create the 2x2 subplot layout ---
# fig: The entire figure
# axs: A 2x2 array of axes objects, representing each subplot
fig, axs = plt.subplots(2, 2, figsize=(10, 8)) # figsize sets the figure size (width, height) in inches
fig.suptitle('Various Plot Types in a 2x2 Subplot Layout', fontsize=16)


# --- Subplot 1: Line Chart (Top-Left) ---
axs[0, 0].plot(x_line, y_line, color='blue')
axs[0, 0].set_title('Line Chart (sin(x))')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].grid(True)


# --- Subplot 2: Bar Chart (Top-Right) ---
axs[0, 1].bar(categories_bar, values_bar, color=colors_bar)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')


# --- Subplot 3: Scatter Plot (Bottom-Left) ---
# s for size, c for color
axs[1, 0].scatter(x_scatter, y_scatter, s=sizes_scatter, c=colors_scatter, cmap='viridis', alpha=0.7)
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')
axs[1, 0].grid(True)


# --- Subplot 4: Pie Chart (Bottom-Right) ---
axs[1, 1].pie(sizes_pie, labels=labels_pie, colors=colors_pie, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.


# --- Adjust layout and display ---
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent titles from overlapping
plt.show()
```
