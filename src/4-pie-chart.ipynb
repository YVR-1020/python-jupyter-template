### Exercise 4: Pie Chart with Percentages

```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript', 'Other'] # Completed labels
sizes = [40, 25, 15, 10, 10] # Example sizes (must sum to 100 or represent proportions)

# Define a list of colors for the slices (optional, but good for clarity)
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'plum']

# Create the pie chart
# autopct='%1.1f%%' formats the percentage to one decimal place
# startangle=90 rotates the start of the first slice
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Add a title to the chart
plt.title("Programming Language Popularity")

# Ensure the circle is drawn as a circle (equal aspect ratio)
plt.axis('equal')

# Show the chart
plt.show()
```
