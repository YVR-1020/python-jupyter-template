### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values
x = np.linspace(0, 10, 100)  # Creates 100 points between 0 and 10

# Define y values (example: y = 2*x + 1)
y = 2*x + 1

# Create the plot
plt.plot(x, y)

# Customize the plot
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)

# Show the plot
plt.show()
```
